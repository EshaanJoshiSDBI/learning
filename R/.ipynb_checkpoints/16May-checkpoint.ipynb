```r
library(dplyr)
library(readr)
library(ggplot2)
```

```r
emp_sim = read_csv('emp_sim.csv')
```

```r
head(emp_sim)
```

```r
qplot(extrovert,revenue, data = emp_sim)
```

```r
qplot(extrovert,revenue, data = emp_sim, geom = 'boxplot')
```

```r
qplot(extrovert,revenue, data = emp_sim, geom = 'violin')
```

Summary Stats

```r
emp_sim %>% summarize(avg = mean(revenue))
```

```r
emp_sim %>% summarize(avg = mean(revenue), med = median(revenue), count = n())
```

```r
emp_sim %>% group_by(education) %>% summarize(avg = mean(revenue), med = median(revenue), count = n())
```

Subsetting using filter

```r
no_prior = emp_sim %>% filter(prior_sales == 0)
```

```r
head(no_prior)
```

```r
q10 = quantile(emp_sim$revenue, 0.1) # bottom 10%
q10
```

```r
q90 = quantile(emp_sim$revenue, 0.9) # 90 percentile / top 10%
q90
```

```r
top10 = emp_sim %>% filter(revenue >= q90)
bottom10 = emp_sim %>% filter(revenue <= q10)
```

```r
summary(top10)
```

```r
summary(bottom10)
```

Factors

```r
ext_levels = c('LOW','MED','HIGH')
emp_sim = emp_sim %>% mutate(ext_factors = factor(extrovert, levels = ext_levels))
```

```r
glimpse(emp_sim)
```

```r
top10 = emp_sim %>% filter(revenue >= q90)
bottom10 = emp_sim %>% filter(revenue <= q10)
```

```r
summary(top10)
```

```r
summary(bottom10)
```

```r
qplot(ext_factors,revenue, data = emp_sim, geom = 'violin')
```
